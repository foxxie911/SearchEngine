# Task 3 (Bonus)

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords') 
```

```python
bad_words = set(stopwords.words('english'))
punctuations = set(string.punctuation)
content_df['Filtered Content'] = ''
for index,row in content_df.iterrows():
    tokens = word_tokenize(row['Content'].lower())
    filtered_content = [ word for word in tokens if 
                        word not in bad_words
                        and word not in punctuations
                       ]
    content_df.at[index, 'Filtered Content'] = filtered_content
content_df
```

# Task 4

```python
web_content = dict(zip(content_df["URL"], content_df["Filtered Content"]))
```

```python
print("\nWebsite Content:")
for site, content in web_content.items():
    print(f"{site}: {content}")
```

```python
def inverted_index(web_content):
    inverted_index = {}
    for url, filtered_content in web_content.items():
        for word in filtered_content:
            if word not in inverted_index:
                inverted_index[word] = []
            inverted_index[word].append(url)
    return inverted_index
```

```python
index = inverted_index(web_content)
print("\nInverted Index:")
for word, websites in index.items():
    print(f"{word}: {websites}")
```

# Task 5

```python
hubs, authorities = nx.hits(graph, max_iter=100, tol=1e-6)
print("\nHITS - Hub Scores:", hubs)
print("\nHITS - Authority Scores:", authorities)
```

```python
pagerank_scores = nx.pagerank(graph, alpha=0.85, max_iter=100, tol=1e-6)
print("\nPageRank Scores:", pagerank_scores)
```

```python
def search_engine(query, index, scores):
    query_terms = query.lower().split()
    results = set()
    for term in query_terms:
        if term in index:
            if not results:
                results = set(index[term])
            else:
                results = results.intersection(index[term])

    # Sort results based on score
    ranked_results = []
    for website in results:
        if website in scores:
          ranked_results.append((website, scores[website]))
    ranked_results.sort(key=lambda x: x[1], reverse=True)

    return ranked_results
```

```python
query = "data python"
print(f"\n##############################Search Results for '{query}' using PageRank#####################################")
results = search_engine(query, index, pagerank_scores)
for page, score in results:
    content_obj = content_df.loc[content_df['URL'] == page, 'Content']
    content = content_obj.iloc[0]
    print(f"{page}: {content} ({score})")

print(f"\n##############################Search Results for '{query}' using HITS (Authorities)############################")
results = search_engine(query, index, authorities)
for page, score in results:
    content_obj = content_df.loc[content_df['URL'] == page, 'Content']
    content = content_obj.iloc[0]
    print(f"{page}: {content} ({score})")
```

# Task 6 (Bonus)

```python
def search_engine_extended(query, index, scores):
    query_terms = query.lower().split()
    results = set()
    for term in query_terms:
        if term in index:
            results = results.union(index[term])

    # Sort results based on score
    ranked_results = []
    for website in results:
        if website in scores:
          ranked_results.append((website, scores[website]))
    ranked_results.sort(key=lambda x: x[1], reverse=True)

    return ranked_results
```

```python
query = "data python"
print(f"\n#################################Search Results for '{query}' using PageRank#####################################")
results = search_engine_extended(query, index, pagerank_scores)

for page, score in results:
    content_obj = content_df.loc[content_df['URL'] == page, 'Content']
    content = content_obj.iloc[0]
    print(f"{page}: {content} ({score})")

print(f"\n################################Search Results for '{query}' using HITS (Authorities)############################")
results = search_engine(query, index, authorities)
for page, score in results:
    content_obj = content_df.loc[content_df['URL'] == page, 'Content']
    content = content_obj.iloc[0]
    print(f"{page}: {content} ({score})")
```
